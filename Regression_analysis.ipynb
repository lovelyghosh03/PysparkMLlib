```python
import pyspark
```

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName("ML").getOrCreate()
```

```python
df=spark.read.csv('boston1.csv',header=True,inferSchema=True)
```

```python
df.show(5)
```

```python
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
```

```python
x.columns
```

```python
assmblr=VectorAssembler(inputCols=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'],outputCol='features')
```

```python
output=assmblr.transform(df)
```

```python
final_df=output.select('features','MEDV')
final_df.show()
```

```python
train_data,test_data=final_df.randomSplit([0.8,0.2])
train_data.show()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
lr=LinearRegression(labelCol='MEDV')
```

```python
model=lr.fit(train_data)
pred=model.evaluate(test_data)
```

```python
pred.predictions.show(5)
```

```python
pred.rootMeanSquaredError
```
